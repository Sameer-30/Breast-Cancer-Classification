```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing


```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)

```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rowss of the dataframe
data_frame.head()
```

```python
# adding the target columns to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print the last 5 rows of the datafrfame
data_frame.tail()
```

```python
# number of rows and column in the dataset
data_frame.shape
```

```python
# Generating the information about the dataset
data_frame.info()
```

```python
# checking for missing value
data_frame.isnull().sum()
```

```python
# statistical measure about the dataset
data_frame.describe()
```

```python
# checking the distribution of Target variable
data_frame['label'].value_counts()
```

```python


data_frame.groupby('label').mean()
```

```python
# Separating teh fratures abd target 
X = data_frame.drop(columns= 'label', axis =1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting the data into training data & testing dta
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training
# Logistic Regression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
# Accuracy Score
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```
